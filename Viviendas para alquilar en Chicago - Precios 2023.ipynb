* # REGRESIÓN LINEAL (con grupo de entrenamiento y test)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Divide los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
### creamos y entrenamos el modelo
model = LinearRegression()
model.fit(X_train , y_train)
```

```python
### Creamos una prediccion con los siguientes datos
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

##### Resultados
print("Error cuadrático medio (MSE):", mse)
print("Coeficiente de determinación (R2 score):", r2)
```

* ## Obtenemos casi el mismo resultado = %15.48. En si empeoro un poquito a comparación de la primera regresión.
    * ### Ahora camibaremos de modelo
----

* # SVR (support vector regressión) con valores normalizados.

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Normalizamos los datos para poder hacer el preprocesado
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
### KERNEL de tipo radial "RBF"
regressor = SVR(kernel="rbf", gamma = 0.1)
regressor.fit(X,y)
```

```python
#### Ahora en este cambiamos el kernel a tipo polinomio 
regressor_polinomio = SVR(kernel="poly", degree = 3)
regressor_polinomio.fit(X,y)
```

```python
regressor.score(X,y)*100
```

```python
regressor_polinomio.score(X,y)*100
```

* # OBSERVACIÓN DEL SVR CON LOS KERNEL "rbf y poly".
    * ## Vemos una mejora muy buena con la regresion lineal Obtenemos unos resultados bastante buenos. Un SVR de %56.30 con el kernel "POLY" .
    * ## Pero aun mejor con el KERNEL "RBF" con un resultado del %77,84

* # SVR con kernel "rbf" (grupo de entrenamiento y test)

```python
x_train , x_test , y_train , y_test = train_test_split(X,y , test_size=0.3 , random_state=42)

#Entrenamos el modelo en los datos de entrenamiento
regressor.fit(x_train , y_train)
```

```python
score_test = regressor.score(x_test , y_test) #Evalúamos el modelo con sus datos de prueba 
```

```python
print(f"El score en los datos de prueba es de : % {score_test*100:.2f}")
```

```python
####Evalúamos el modelo en los datos de entrenamiento y vemos su resultado.
score_train = regressor.score(x_train , y_train) 
print(f"Eficacia del modelo con los datos de entrenamiento son de: %{score_train*100:.2f}")
```

* # Hacemos una validación cruzada para ver como se comporta el modelo (Cross validation)

```python
from sklearn.model_selection import cross_val_score
```

```python
#Aplivamos la validacion cruzada con un fold de 5
scores = cross_val_score(regressor , X,y , cv=10)

print("Eficacia del modelo haciendo la validación cruzada en cada fold son : %",scores)

print("El score promedio en la validación cruzada es de:",scores.mean()*100)
```

* ### El resultado con el cross validation sigue siendo bastante bueno, vemos una mejora grande comparando con los primero modelos creados.
---

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
# Predeciremos los valores para los datos de prueba
y_pred = regressor.predict(x_test)

#Calculamos el error cuadratico medio (MSE)
mse = mean_squared_error(y_test , y_pred)
mae = mean_absolute_error(y_test , y_pred)

print("El error cuadratico medio (MSE) es de :", mse)
print("El error absoluto medio (MAE) es de: ", mae)
```

* ## Podemos ver que tanto el MSE y MAE los resultados que obtuvimos con el modelo SVR indican un buen rendimiento, ya que tanto el MSE como el MAE son bajos, es decir mas bajo o cercano al cero mejor es el resultado.

* ## El MAE y el MSE son medidas comunes para evaluar el rendimiento de un modelo y que se utilizan para entender cuán cerca están las predicciones del modelo de los valores reales
    * #### Error Absoluto Medio (MAE): Es una medida del error promedio entre las predicciones de un modelo y los valores reales. 
    * #### Error Cuadrático Medio (MSE): El MSE también es una medida del error entre las predicciones y los valores reales, pero en lugar de tomar la diferencia absoluta, toma la diferencia al cuadrado. Esto significa que los errores más grandes se amplifican más que los errores más pequeños

* # ÁRBOL DE DECISIÓN 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
##### Divide los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
##### modelo de árbol de decisión
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
#### predicciones en los datos de prueba
y_pred = model.predict(X_test)

#### Ver precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo:", accuracy*100)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de confusión:")
print(confusion_matrix)
```

* Podemos ver que solamente el modelo cometio 2 errores en total, hubo una falsa predicción de clase 0 y una falsa predicción de clase 1. 
----

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
### Veremos otras metricas para evaluar el rendimiento del modelo
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precisión: {precision*100:.2f}")
print(f"Recall: {recall*100:.2f}")
print(f"F1-score: {f1*100:.2f}")
```

---
* ### Validación cruzada para ver como se comporta el modelo

```python
scores = cross_val_score(model, X, y, cv=6)
print("Precisión promedio en validación cruzada:", scores.mean()*100)
```
